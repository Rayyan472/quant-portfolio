```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

tickers = ["TSLA", "KO"]
data = yf.download(tickers, start="2020-01-01", end="2025-01-01")["Close"]

data.columns = tickers

daily_returns = data.pct_change()

cumulative_returns = (data / data.iloc[0])
cumulative_returns = cumulative_returns - 1

volatility = daily_returns.std()

for ticker in tickers:
    print(f"{ticker} Cumulative Return: {cumulative_returns[ticker].iloc[-1]:.2%}")
    print(f"{ticker}Daily Volatility: {volatility[ticker]:.2%}")
    print("-" * 30)
plt.figure(figsize = (10, 6))
for ticker in tickers:
    plt.plot(cumulative_returns.index, cumulative_returns[ticker], label = ticker)
plt.title("Tesla vs KO Cumulative Return")
plt.xlabel("Date")
plt.ylabel("Cumulative Return (%)")
plt.legend()
plt.grid(True)
plt.show()
```
