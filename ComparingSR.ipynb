```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

tickers = ["TSLA", "KO", "AAPL"]
data = yf.download(tickers, start="2020-01-01", end="2025-01-01")["Close"]
data.columns = tickers

normalized_data = data/data.iloc[0]

daily_returns = data.pct_change()

sharpe_ratio = (daily_returns.mean()/daily_returns.std()) * (252**0.5)

plt.figure(figsize=(10,6))
for ticker in tickers:
    plt.plot(normalized_data.index, normalized_data[ticker] - 1, label=ticker)

plt.title("Cumulative Returns Comparison (2020-2025)")
plt.xlabel("Date")
plt.ylabel("Cumulative Returns (%)")
plt.legend()
plt.grid(True)
plt.show()

print("Sharpe Ratio (Risk-Adjusted Returns):")
for ticker in tickers:
        print(f"{ticker}: {sharpe_ratio[ticker]:.2f}")
```
